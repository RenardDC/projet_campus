#COMBIEN D'ARBRES ONT ETE PLANTER ANNEE PAR ANNEE (ex : 1987 : 771, 1988 : 266, etc…) ?

```python
compte_elements = {}
for element in annees_plantation:
    if element in compte_elements:
        compte_elements[element] += 1
    else:
        compte_elements[element] = 1

for element in sorted(compte_elements):
    print(f"{element}:{compte_elements[element]}")
```

#COMMENT LA PRATIQUE DE PLANTATION A T ELLE EVOLUER AU COURS DU TEMPS, DIVERSIFIE T-ON PLUS QU'AVANT?

```python
liste_annee_f = []
for ligne in data[1:]:
    if ligne[19]:
        liste_annee_f.append( ligne[19])
liste_annee_f
```

```python
liste_genre_espece_f = []
for ligne in data [1:]:
    genre = ligne[13]
    espece = ligne [14]
    if genre and espece:
        espece_full = (f"{genre} : {espece}")
        liste_genre_espece_f.append(espece_full)
liste_genre_espece_f_unique = set(liste_genre_espece_f)
```

```python
liste_finale_f = list(zip(liste_annee_f, (liste_genre_espece_f)))
liste_finale_f
```

```python
compte_final_diversiter = {}
for annee, genre_espece in liste_finale_f:
    if annee not in compte_final_diversiter:
        compte_final_diversiter[annee] = set()
    compte_final_diversiter[annee].add((genre_espece))
nb_espece_annee = {annee: len(genre_espece) for annee, genre_espece in compte_final_diversiter.items()}
nb_espece_annee_trier = sorted(nb_espece_annee.items(), key=lambda x: x[0])
nb_espece_annee_trier
```

```python
liste_genre_espece_f_unique = set(liste_genre_espece_f)
liste_finale_f2 = list(zip(liste_annee_f, (liste_genre_espece_f_unique)))
liste_finale_f2
```

```python
compte_unique_f = {}
for annee, genre_espece in liste_finale_f2:
    if annee not in compte_unique_f:
        compte_unique_f[annee] = set()
    compte_unique_f[annee].add((genre_espece))
nb_espece_anneef = {annee: len(genre_espece) for annee, genre_espece in compte_unique_f.items()}
nb_espece_annee_trierf = sorted(nb_espece_anneef.items(), key=lambda x: x[0])
nb_espece_annee_trierf
```

```python
annees = [int(item[0]) for item in nb_espece_annee_trier]
nb_espece = [item[1] for item in nb_espece_annee_trier]

plt.bar(annees, nb_espece, color='orange', width=0.8)

plt.xlabel("Année", fontsize=20)
plt.ylabel("Nombre d'arbres plantés", fontsize=20)
plt.title("Nombre d'espèces diffèrentes plantées par année dans la ville de Grenoble", fontsize=30)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True)
plt.subplots_adjust(left=0.1, right=2.5)
plt.show()
```
