# Compter le nombre de représentants de chaque genre botanique

```python
genre_counts = df['genre_bota'].value_counts().to_dict() 
```

============================================================================================================================================================

# Trouver le genre botanique le plus représenté 

```python
genre_max_representation = max(genre_counts, key=genre_counts.get)
```

============================================================================================================================================================

 # Liste des représentants uniques de leur genre botanique 

```python
unique_representatives = df.drop_duplicates(subset=['genre_bota']) 
unique_representatives
```

============================================================================================================================================================

# Classer tous les genres botaniques par ordre décroissant de nombre de représentants

```python
sorted_genre_counts = dict(sorted(genre_counts.items(), key=lambda item: item[1], reverse=True))
```

============================================================================================================================================================

# Proportion du genre botanique le plus représenté

```python
total_arbres = df.shape[0] 
proportion_max_genre = genre_counts[genre_max_representation] / total_arbres
proportion_max_genre
```

============================================================================================================================================================

 # Proportion de chaque genre botanique

 # Analyse de la diversité des plantations au fil du temps 

```python
diversity_over_time = df.groupby(['anneedeplantation', 'genre_bota']).size().unstack(fill_value=0) 
```

============================================================================================================================================================

# Diversité par décennie 

```python
diversity_by_decade = df.copy() 
diversity_by_decade['decade'] = (diversity_by_decade['anneedeplantation'] // 10) * 10 
diversity_by_decade = diversity_by_decade.groupby(['decade', 'genre_bota']).size().unstack(fill_value=0)
```
