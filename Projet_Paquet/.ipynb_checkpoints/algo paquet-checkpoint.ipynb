```python
import random

types = ['Coeur', 'Carreau', 'Trefle', 'Pique']
valeurs = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Valet', 'Dame', 'Roi', 'As']

paquet = [(type, valeur) for type in types for valeur in valeurs]

paquet.append(('Joker', 'Rouge'))
paquet.append(('Joker', 'Noir'))

random.shuffle(paquet)
```

```python
paquet
```

```python
liste_coeur = []
liste_carreau = []
liste_trefle = []
liste_pique = []
liste_joker = []

def tri_type_liste(carte):
    for carte in paquet:
        type = carte[0]
        if type == 'Coeur':
            liste_coeur.append(carte)
        elif type == 'Carreau':
            liste_carreau.append(carte)
        elif type == 'Trefle':
            liste_trefle.append(carte)
        elif type == 'Pique':
            liste_pique.append(carte)
        else:
            liste_joker.append(carte)
    liste_coeur + liste_carreau + liste_pique + liste_trefle + liste_joker
    return liste_coeur, liste_carreau, liste_pique, liste_trefle, liste_joker
tri_type_liste(paquet)
```

```python
tri_type_liste(paquet)
```

```python
def tri_liste(liste):
    valeurs = {
        "As": 0,
        "2": 1,
        "3": 2,
        "4": 3,
        "5": 4,
        "6": 5,
        "7": 6,
        "8": 7,
        "9": 8,
        "10": 9,
        "Valet": 10,
        "Dame": 11,
        "Roi": 12
    }

    liste_triee = []

    liste_check = set()

    for carte in liste:
        type_carte, valeur = carte
        valeur_carte = valeurs[valeur]

        if carte not in liste_check:
            position = 0
            while position < len(liste_triee) and valeurs[liste_triee[position][1]] < valeur_carte:
                position += 1

            liste_triee.insert(position, carte)
            liste_check.add(carte)

    return liste_triee
tri_liste(liste_coeur)
```

```python
paquet_trier_final = tri_liste(liste_coeur) + tri_liste(liste_carreau) + tri_liste(liste_pique) + tri_liste(liste_trefle) + liste_joker
```
